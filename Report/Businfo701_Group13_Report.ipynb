## Data Source and Audit

### Data Source
The "Greenhouse Gas Emissions Inventories" dataset is provided by **OECD** as part of their Air Emission data list. This dataset records the total greenhouse gas emissions of different countries and regions, categorized by year (in thousand tonnes of CO2 equivalent).  **OECD** is an international organization for economic cooperation and development, among the most significant and most trustworthy sources of globally comparable environmental and socio-economic data in the world
The data is provided in **CSV format**, separated by commas, making it easy to load and analyze.  
For the project, we selected data from **2011 to 2021**, as this period offers a complete and comprehensive range, supporting multi-year trend analysis and cross-country comparisons. The dataset includes greenhouse gas emissions data for **38 OECD member countries**, **19 non-member countries**, and **5 OECD regions**.

### Data Characteristics
1. **Number of Records and Columns**  
 The dataset is in an extended format, containing **674 records** and **30 columns**. Each record corresponds to the emissions data of a specific country or region for a particular year.

2. **Key Variables**
   - **`Reference area`**: Name of the country or region.
   - **`REF_AREA`**: Code of the country or region.
   - **`POLLUTANT`** and **`Pollutant`**: Name of the pollutant (full name and abbreviation).
   - **`TIME_PERIOD`**: Observation year.
   - **`OBS_VALUE`**: Greenhouse gas emissions, measured in  tonnes of CO2 equivalent.  
 Other columns include data flow identifiers, measurement types (e.g., **`MEASURE`**), unit conversion information, and more.

An initial audit of the dataset revealed the following issues that were addressed during data cleaning:

1. **Unused Columns**  
 Some columns have no practical meaning or are empty, such as **`PRICE_BASE`** and **`CONVERSION_TYPE`**, which were removed during processing.

2. **Shifted Data**  
 The first row of the dataset has shifted downward, causing a mismatch between the header and the actual data. These erroneous rows are irrelevant for analysis and should be removed.

3. **Duplicate Columns**  
 The dataset contains duplicate or redundant information, such as **`Pollutant`** (full pollutant name) and **`POLLUTANT`** (abbreviated name). Only one of these columns was retained.

4. **Missing Data**  
 Although the **`OBS_VALUE`** column (observed values) did not show explicit missing data, an inspection of the **`TIME_PERIOD`** column revealed that some countries lack complete data for all ten years. This could be due to rows being removed with empty years during data download, resulting in implicit missing data.

1. **Determine Primary Keys**  
 By examining the characteristics of other datasets, we will identify the primary keys that support data merging.

# Selected Columns and Reasons for Retention

### Retained Columns and Justifications
For the project, six columns were selected : **`REF_AREA`**, **`Reference area`**, **`POLLUTANT`**, **`TIME_PERIOD`**, **`OBS_VALUE`**, and **`Measure`**. The reasons for retaining each column are as follows:

1. **`REF_AREA`**  
 This column contains three-character country codes, which are retained as a primary key for merging datasets. Country codes are more standardized and accurate than country names, as country names might vary across different databases, whereas codes are uniform, ensuring precise identification of countries.

2. **`TIME_PERIOD`**  
 This column represents the observation year and is retained as another primary key for merging datasets. The range of 2011–2021 allows us to identify trends in greenhouse gas emissions across different countries.

3. **`Reference area`**  
 This column provides the full name of the country or region and is retained for easy identification. While country codes are helpful for data processing, they could be more intuitive for general audiences. Retaining this column makes the data and visualizations more understandable to viewers.

4. **`POLLUTANT`**  
 This column specifies the name of the pollutant and is retained to identify pollutant types. Additional datasets may include various pollutant categories, so this column ensures compatibility and supports diverse analyses.

5. **`OBS_VALUE`**  
 This column contains the observed values, representing the total greenhouse gas emissions. It is a core indicator for analysis and directly reflects emissions levels across countries and years, making it essential for our project.

6. **`Measure`**  
 This column describes the measurement of greenhouse gases (GHG), excluding emissions or removals resulting from land use, land-use change, and forestry (LULUCF). Excluding LULUCF ensures the data more accurately reflects the direct impact of human activities, such as industry, energy, and transportation, without interference from natural or land-use factors. This column is retained to provide context for the scope and background of the data.

To ensure data consistency and ease of use, we selected the **CSV format**, with data separated by commas.

The dataset covers the same countries and years as the first dataset (2011–2021), including **38 OECD member countries**, **19 non-member countries**, and **5 OECD regions**. The dataset contains **30 columns** and **6,793 records**, with fields covering country names and codes, data flow identifiers, pollutant types, measurement methods, years, and observed values. The data is in a **long format** and requires cleaning and transformation to meet project requirements.

Since the structure of this dataset is similar to the OECD additional pollutant dataset, the data-cleaning process is essentially the same. The main steps include:

````python
#importing different industry
OECD_pollutant_different_industry = pd.read_csv("different_industry.csv")
OECD_pollutant_different_industry = OECD_pollutant_different_industry.loc[:,["REF_AREA", "Reference area", "POLLUTANT", "TIME_PERIOD", "OBS_VALUE", "UNIT_MEASURE", "Measure"]]
OECD_pollutant_different_industry = OECD_pollutant_different_industry.loc[(OECD_pollutant_different_industry["Reference area"].isin(country_list)) & (OECD_pollutant_different_industry["UNIT_MEASURE"] == "T_CO2E")]
OECD_pollutant_different_industry = OECD_pollutant_different_industry.sort_values(["Reference area", "TIME_PERIOD"])
OECD_pollutant_different_industry.reset_index(inplace = True, drop = True)


OECD_pollutant_different_industry
````

# Data Merging

We cleaned and transformed the three OECD datasets (**Original Greenhouse Gas Data**, **Additional Pollutant Data**, and **Greenhouse Gas Source Data**) into a pre-defined **standardized structure** and merged them into a single comprehensive dataset.

## Standardized Structure

The core of the **standardized structure** includes:

- **Primary Key Columns**: Country code (`REF_AREA`) and year (`TIME_PERIOD`) as unique identifiers for each record.
- **Auxiliary Column**: Country name (`Reference area`), to help non-technical users recognize countries and understand visualizations.
- **Indicator Columns**: Each pollutant and its measurement description are stored as independent columns containing the corresponding observed values.

This structure ensured consistency across the three datasets before merging, facilitating unified operations and analysis.



## Merging Process

### 1. Merging Operations
Using the `merge` function, we merged the three datasets step by step, based on the primary key columns **`REF_AREA`** (Country Code) and **`TIME_PERIOD`** (Year):

- **Step 1**: Merged **`OECD_original_data`** (Original Greenhouse Gas Data) with **`OECD_other_pollutants`** (Additional Pollutant Data).
- **Step 2**: Merged the result with **`OECD_pollutant_different_industry`** (Greenhouse Gas Source Data), ensuring all indicators aligned in a single table.

### 2. Index Resetting
After merging, the index was reset using `reset_index` to simplify the structure and remove unnecessary index columns, making the data more organized.

### 3. Column Renaming
Column names were adjusted to improve readability.

- **`REF_AREA`** was renamed to **`Country Code`**, making it easier for both technical and non-technical users to understand the purpose of the field.
- **`Reference area_x`** was renamed to **`Country Name`**, allowing users to identify the country, especially for visualizations quickly.



## Final Dataset

The merged dataset, **`OECD_data_all`**, contains **332 rows** and **25 columns**, integrating all relevant greenhouse gas emission indicators, pollutant types, and source industry data. This comprehensive structure provides a strong foundation for further analysis and visualization.




````python
#Meringing the three datasets
OECD_data_all = pd.merge(OECD_original_data, OECD_other_pollutants).merge(OECD_pollutant_different_industry, on=["REF_AREA", "TIME_PERIOD"])
OECD_data_all.reset_index(inplace=True, drop=True)
OECD_data_all = OECD_data_all.rename(columns={"REF_AREA": "Country Code",
                                    "Reference area_x": "Country Name"})

OECD_data_all
````

## Data Source

The dataset contains **18 columns** and **6,655 rows**, including country names and codes, indicator names and codes, and annual observations from 2009 to 2022. The original data is in a **wide format**, which requires cleaning and transformation to meet analysis needs.



## Initial Audit
An initial audit of the dataset revealed the following issues to be addressed:

1. **Redundant Information**  
 The dataset includes irrelevant information at the bottom, such as download dates and update records. These need to be removed during the cleaning process to ensure data integrity.

2. **Missing Values**  
 There are numerous missing values in the dataset, likely due to a lack of observations or difficulties in estimating specific values for some countries and regions. At this stage, we cannot impute or delete all missing values. Further data cleaning and filtering will determine how to handle these missing values based on analysis needs.

3. **Inconsistent Country Names**  
 Country names in the World Bank dataset do not always match those in the OECD dataset. For example, "Türkiye" in the OECD dataset appears as "Turkiye" in the World Bank dataset. However, since we use **country codes and years as primary keys** for data merging, this inconsistency does not affect the merging process.

4. **Time Field Format Issues**  
 The time field in the World Bank dataset (e.g., `"2010 [YR2010]"`) includes extra information that complicates merging and analysis. This field needs to be cleaned and transformed to extract pure year values.



## Cleaning Steps

### 1. Data Import and Initial Processing  
- Used the `pd.read_csv` function to load the data and extracted the first **6,649 rows**, excluding redundant information at the bottom of the file to retain only relevant data.

### 2. Data Format Transformation  
- **Applied the `melt` function to convert the data from wide format to long format**:  
  - **`id_vars`**: Retained columns that do not vary by year, including **`Series Name`** (indicator name), **`Series Code`** (indicator code), **`Country Name`**, and **`Country Code`**.  
  - **`value_vars`**: Transformed annual observations (e.g., `"2009 [YR2009]"` to `"2022 [YR2022]"`) into a single column, creating a unified time series.  
  - **Result**: Generated a long-format dataset with **`TIME_PERIOD`** (year) and **`OBS_VALUE`** (observation value) columns, facilitating time-series analysis.

### 3. Time Extraction and Cleaning  
- The time field includes extra information (e.g., `"2010 [YR2010]"`) that affects primary key merging. We addressed this issue by:  
  - Defining the `get_year` function to split the time field and extract the year (e.g., `"2010 [YR2010]"` → `"2010"`).  
  - Applying the function to clean the **`TIME_PERIOD`** column, ensuring that year values are consistent and formatted as integers.

### 4. Column Optimization and Deletion  
- Deleted the original **`Series Code`** column, as it is redundant and does not impact the analysis.  
- Retained the following key columns:  
  - **`Series Name`**: Indicator name for identifying specific economic or environmental metrics.  
  - **`Country Name`** and **`Country Code`**: Used to identify countries and facilitate data visualization.  
  - **`TIME_PERIOD`**: Year column, standardized as a unique time identifier.  
  - **`OBS_VALUE`**: Observation value column, recording quantitative metrics.

````python
# Importing the necessary modules:
from cpdb_api import request  # 'request' helps us interact with the CPDB API to fetch policy data
import pandas as pd  # 'pandas' helps us work with the data in tabular format (DataFrames)
````

3. Containing a country code and an empty set.
- ```country_codes = ["NZL"]``` Defining the list of country codes for which we want to collect policy data. 
- ```all_policy_data = []``` Creating an empty list to store the policy data as we fetch it in the next step using loop.


````python
# country_code will collect data for New Zealand, represented by the country code "NZL"
country_codes = ["NZL"]
# Creating an empty list to store the policy data as we fetch it
all_policy_data = []
````

4. Create a loop and web scrapping

- ```for year in range(2011, 2022):``` Creating a to go through each country in the list. Add inner loop which filters the country data by year.

- ```request.Request()``` This line creates a new Request object, which is used to send queries to the CPDB API.

- ```r.set_country(country_code)``` This line sets the filter for the country code in the request to ensure that only data for the specified country is fetched.

- ```r.set_decision_date(year)``` This line sets the filter for the decision year of the policy in the request.

- ```policy_df = r.issue()``` This line sends the configured request to the CPDB API and retrieves the policy data. The 'issue' method returns the data in a pandas DataFrame as per documentation.

- ```all_policy_data.append(policy_df)``` This line appends the fetched data (as a pandas DataFrame) to the all_policy_data list.

- ```final_policy_df = pd.concat(all_policy_data, ignore_index=True)``` This line combines all the individual DataFrames in the all_policy_data list into one final DataFrame (final_policy_df).

````python
# Starting a loop to go through each country in the list 'country_codes'
for country_code in country_codes:
    
    # Looping through the years from 2011 to 2021 (inclusive)
    for year in range(2011, 2022):  # 2021 is included in the range, so the loop will end at 2021
        
        # Creating a new 'Request' object to make an API call for the given country and year
        r = request.Request()  # This object will help us set filters for the data we want
        
        # Setting the country filter for the request to the current country code (e.g., "NZL")
        r.set_country(country_code)
        
        # Setting the year filter for the request to the current year (e.g., 2011, 2012, etc.)
        r.set_decision_date(year)
        
        # Issuing the request to fetch the data. The 'issue' method returns the data in a pandas DataFrame
        policy_df = r.issue()
        
        # Appending the fetched data (as a DataFrame) to our list 'all_policy_data'
        all_policy_data.append(policy_df)

# After the loop ends, we combine all the policy data that we've gathered into one large DataFrame
# 'pd.concat' is used to concatenate the list of DataFrames into a single DataFrame.
final_policy_df = pd.concat(all_policy_data, ignore_index=True)

# Now, 'final_policy_df' contains the combined data for all countries and years we requested
# Finally, we display the resulting DataFrame with all the policy data
final_policy_df.head(5)
````

5. Selecting required variables from the whole dataset.
- ```final_policy_df.iloc[:,[1, 2, 4, 6, 9, 10, 11, 12, 14, 15, 19, 20]]``` Selecting Variables using Column index.

````python
# Selecting variables from final_policy dataframe and storing it into final_crawled_sample dataframe
final_crawled_sample = final_policy_df.iloc[:,[1, 2, 3, 6, 9, 10, 11, 12, 14, 15, 19, 20]]
final_crawled_sample
````
